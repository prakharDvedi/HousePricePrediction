```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1,2,3,4,5,5.20,5.70,5.80,7,10])
y_train=np.array([5,9.8,14.7,21,25.5,27.2,28.4,29.3,38,54.2])
```

```python
def cost_function(x,y,w,b):
    m=x.shape[0]
    cost=0
    for i in range(m):
        f_wb=w*x[i]+b
        cost=cost+(f_wb-y)**2
    total_cost=1/(2*m)*cost
    return total_cost
```

```python
def compute_gradient(x,y,w,b):
    m=x.shape[0]
    dj_dw=0
    dj_db=0
    for i in range(m):
        f_wb=w*x[i]+b
        dj_dw_i=(f_wb-y[i])*x[i]
        dj_db_i=(f_wb-y[i])
        dj_dw+=dj_dw_i
        dj_db+=dj_db_i
    dj_db=dj_db/m
    dj_dw=dj_dw/m
    return dj_dw,dj_db
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,num_iters):
    j_history=[]
    p_history=[]
    b=b_in
    w=w_in
    for i in range(num_iters):
        dj_dw,dj_db=compute_gradient(x,y,w,b)
        b=b-alpha*dj_db
        w=w-alpha*dj_dw
        

    return w,b
```

```python
w_in=0
b_in=0
iterations=10000
alpha=1.0e-2
compute_cost=cost_function(x_train,y_train,w_in,b_in)
w_final,b_final=gradient_descent(x_train,y_train,w_in,b_in,alpha,iterations)
print(w_final)
print(b_final)
```
